```python
import pandas as pd
```

```python
def extractJobs(path, adjust=0):
    jobs = []
    with open(path) as f:
        data = f.read()[:-1] # do not include "\n"
        for row in data.split("\n"):     # each row in the file
            row_list = row.split(",")    # split each line with ","
            if(len(row_list[0]) != 0):   # if the row has a value
                jobs.append(row_list[1+adjust]) # append the job to jobs list
    return jobs[1-adjust:]

def oldJobs(path,adjust=1):

    with open(path,'w+') as f:
        read_char = f.read(1)
        
        if (not read_char):
            return []
    return extractJobs(path,adjust)

def register(df,old_jobs):
    for i in range(len(df)):
        df["Ticket"] = df.iloc[i]["Name"] in old_jobs # save False if no record of this job found

    return df

def merge(paths):
    jobs = []

    for i in paths:
        jobs += extractJobs(i)

    return [i for i in jobs if (len(i) != 0)]

def concat_df(paths, old_jobs):

    df = pd.DataFrame()

    for path in paths:
        df_i = pd.read_csv(path)
        df_i = register(df_i, old_jobs)
        df = pd.concat([df,df_i],ignore_index=True)
    
    return df
```

```python
old_path = "csv/jobs.csv"
paths = ["csv/coderspace.csv","csv/tcmb.csv"]
```

```python
new_jobs = merge(paths)
```

```python
old_jobs = oldJobs(old_path)
```

```python
df = concat_df(paths, old_jobs)
```
